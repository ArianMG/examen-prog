```python
# Importando librerías

import pandas as pd
import matplotlib.pyplot as plt
# %matplotlib inline
```

```python
# Cargando datos y visualizando los 10 primeros

covid19_csv = 'datos_abiertos_covid19.zip'
df_covid19 = pd.read_csv(covid19_csv)

# pd.options.display.max_columns = None
df = df_covid19.copy()
df.head(10)
```

```python
# Obteniendo información de la base de datos
df.info()
```

```python
# Diccionario

# df_dic_covid19 = pd.read_excel(imm_2020_xls, sheet_name='Diccionario', engine='xlrd')
# df_dic_covid19
```

```python
# Convirtiendo  a un formato de fechas
df['FECHA_INGRESO'] = pd.to_datetime(df['FECHA_INGRESO'], format="%Y-%m-%d")
df
```

```python
# Añadiendo el campo SEMANA
df['SEMANA'] = df['FECHA_INGRESO'].dt.isocalendar().week
df
```

```python
# Añadiendo el campo INGRESO y DEFUNCION para indicar numéricamente si el paciente ingresó a falleció

df['INGRESO'] = df['FECHA_INGRESO'].apply(lambda x: 1)
df['DEFUNCION'] = df['FECHA_DEF'].apply(lambda x: 1 if x != '9999-99-99' else 0)

df
```

```python
# Graficando letalidad de covid 19 por semana en México, Sonora y Hermosillo

mexico = df[df['NACIONALIDAD'] == 1][['SEMANA', 'INGRESO', 'DEFUNCION']].groupby(['SEMANA']).sum().apply(lambda x: x['DEFUNCION'] / x['INGRESO'] * 100, axis=1)
sonora = df[df['ENTIDAD_RES'] == 26][['SEMANA', 'INGRESO', 'DEFUNCION']].groupby(['SEMANA']).sum().apply(lambda x: x['DEFUNCION'] / x['INGRESO'] *100, axis=1)
hermosillo = df[df['MUNICIPIO_RES'] == 30][['SEMANA', 'INGRESO', 'DEFUNCION']].groupby(['SEMANA']).sum().apply(lambda x: x['DEFUNCION'] / x['INGRESO'] * 100, axis=1)

fig, ax = plt.subplots(figsize=(14, 10))
ax = mexico.plot.line(label='México')
ax = sonora.plot.line(label='Sonora')
ax = hermosillo.plot.line(label='Hermosillo')
ax.set_ylim(-0.2, 3)
ax.legend(loc=1)
ax.set_xlabel('Semana')
ax.set_ylabel('Letalidad (%)')
ax.set_title('Letalidad de la COVID19 por semana')

plt.savefig('figura.png', facecolor='w', bbox_inches='tight', dpi=200)
```

```python
# Tabla de ingresados, número de decesos e indice de letalidad por grupo etáreo

etario = pd.DataFrame(columns=['INGRESO', 'DECESOS', 'LETALIDAD'], index=['0-18', '18-30', '30-60','+60'])

for i in etario.index:
    if(i == '0-18'):
        etario.loc[i]['INGRESO'] = df[df['EDAD'] <= 18][['INGRESO']].sum()[0]
        etario.loc[i]['DECESOS'] = df[df['EDAD'] <= 18][['DEFUNCION']].sum()[0]
        etario.loc[i]['LETALIDAD'] = df[df['EDAD'] <= 18][['INGRESO', 'DEFUNCION']].apply(lambda x: x['DEFUNCION'] / x['INGRESO'], axis=1).sum()

    elif(i == '18-30'):
        etario.loc[i]['INGRESO'] = df[(df['EDAD'] > 18) & (df['EDAD'] <= 30)][['INGRESO']].sum()[0]
        etario.loc[i]['DECESOS'] = df[(df['EDAD'] > 18) & (df['EDAD'] <= 30)][['DEFUNCION']].sum()[0]
        etario.loc[i]['LETALIDAD'] = df[(df['EDAD'] > 18) & (df['EDAD'] <= 30)][['INGRESO', 'DEFUNCION']].apply(lambda x: x['DEFUNCION'] / x['INGRESO'], axis=1).sum()

    elif(i == '30-60'):
        etario.loc[i]['INGRESO'] = df[(df['EDAD'] > 30) & (df['EDAD'] <= 60)][['INGRESO']].sum()[0]
        etario.loc[i]['DECESOS'] = df[(df['EDAD'] > 30) & (df['EDAD'] <= 60)][['DEFUNCION']].sum()[0]
        etario.loc[i]['LETALIDAD'] = df[(df['EDAD'] > 30) & (df['EDAD'] <= 60)][['INGRESO', 'DEFUNCION']].apply(lambda x: x['DEFUNCION'] / x['INGRESO'], axis=1).sum()


    elif(i == '+60'):
        etario.loc[i]['INGRESO'] = df[df['EDAD'] > 60][['INGRESO']].sum()[0]
        etario.loc[i]['DECESOS'] = df[df['EDAD'] > 60][['DEFUNCION']].sum()[0]        
        etario.loc[i]['LETALIDAD'] = df[df['EDAD'] > 60][['INGRESO', 'DEFUNCION']].apply(lambda x: x['DEFUNCION'] / x['INGRESO'], axis=1).sum()

etario

```

```python
# Guardando en formato csv.zip
etario.to_csv('file.csv.zip', compression='gzip')
```
